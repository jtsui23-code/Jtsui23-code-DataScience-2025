# ðŸ§¼ Cleaning Dirty Data (Missing Values & Type Fixes)

## ðŸ”¹ LEARNING GOALS:
- Detect and count missing values (`NaN`)
- Fill or drop missing data
- Convert column data types safely
- Understand the difference between `NaN`, `None`, `""`, and type mismatches


### ðŸ§ª 1. Load a Messy Dataset

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", None],
    "Age": ["25", "thirty", 35, np.nan, "40"],
    "Signup Date": ["2022-01-01", "not a date", "2022/03/01", None, "April 5, 2022"],
    "Score": [95.5, None, 88.0, 92.5, ""]
}

df = pd.DataFrame(data)
df
```

### ðŸ§¯ 2. Detecting Missing or Broken Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

### ðŸ§¹ 3. Cleaning Strategy Options

```python
df.fillna({
    "Name": "Unknown",
    "Age": -1,
    "Signup Date": "1970-01-01",
    "Score": 0.0
})
```

```python
df.dropna()
```

### ðŸ§¬ 4. Data Type Fixes

```python
df.dtypes
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")
df.dtypes
```

### ðŸ©¹ 5. Impute (Fill In) Fixed Missing Values

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Score"].fillna(df["Score"].mean(), inplace=True)
```

```python
df["Signup Date"].fillna(df["Signup Date"].min(), inplace=True)
```

```python
df["Name"].fillna("Unknown", inplace=True)
```

### ðŸ¤“ 6. Cleaned Data Review

```python
df.info()
```

```python
df.describe(include="all")
```

### ðŸ§  Mini-Challenge

> ðŸ—‚ Load `"data/survey.csv"` and:
- Identify which columns have missing values
- Use `.isnull().sum()` to get a null report
- Use a mix of `.fillna()`, `.dropna()`, and `pd.to_numeric()` or `pd.to_datetime()` to clean it
- Print a summary with `.info()` and `.describe()`

```python
survey = pd.DataFrame({
    "ParticipantID": [101, 102, 103, 104, 105, 106],
    "Email": ["alice@example.com", "bob@", None, "david@example.com", "", "eve@example.com"],
    "Age": [28, "thirty", 22, np.nan, 35, 40],
    "JoinDate": ["2023-05-01", "2023-06-15", "notadate", "2023-07-20", None, "2023-08-01"],
    "FeedbackScore": [4.5, 5, "Excellent", np.nan, 3.5, 4],
    "Subscribed": ["Yes", "No", None, "Yes", "", "No"]
})

print("New Survey DataFrame:")
print(survey)

```

```python
print("\nMissing values per column:")
print(survey.isnull().sum())

```

```python


survey["Age"] = pd.to_numeric(survey["Age"], errors='coerce')
survey["FeedbackScore"] = pd.to_numeric(survey["FeedbackScore"], errors='coerce')
survey["JoinDate"] = pd.to_datetime(survey["JoinDate"], errors='coerce')
```

```python
print(survey)
print("------------------------------------------------------------------------\n")

print(".info\n")
print(survey.info())
print("------------------------------------------------------------------------\n")

print(".describe")
print(survey.describe())
```
