# ðŸ§® Sorting & Filtering in Pandas

## ðŸ”¹ LEARNING GOALS:
- Sort rows using `.sort_values()`
- Filter data using conditional logic
- Chain multiple filters together
- Use `.query()` for readable filtering


### ðŸ“¥ 1. Load the Dataset

Created my own data because could not find students.csv file

```python
import pandas as pd

# df = pd.read_csv("../../data/students.csv")
student_data = {
    'first_name': [
        'Danielle', 'Angel', 'Joshua', 'Jeffrey', 'Jill',
        'Michael', 'Samantha', 'Brian', 'Olivia', 'Ethan',
        'Sophia', 'Ryan', 'Emma', 'David', 'Grace',
        'Lucas', 'Ava', 'Noah', 'Isabella', 'Mason',
        'Liam', 'Charlotte', 'Benjamin', 'Amelia', 'Elijah',
        'Harper', 'Alexander', 'Mia', 'James', 'Ella'
    ],
    'last_name': [
        'Wood', 'Clark', 'Adams', 'Zuniga', 'Wong',
        'Martinez', 'Taylor', 'Anderson', 'Nguyen', 'Hernandez',
        'Moore', 'King', 'Lopez', 'Scott', 'Green',
        'Turner', 'Hall', 'Allen', 'Baker', 'Nelson',
        'Carter', 'Mitchell', 'Perez', 'Roberts', 'Phillips',
        'Campbell', 'Evans', 'Parker', 'Murphy', 'Reed'
    ],
    'math_score': [
        100, 67, 61, 77, 75,
        89, 94, 72, 85, 91,
        68, 80, 97, 73, 84,
        90, 66, 79, 88, 92,
        76, 83, 95, 70, 87,
        64, 98, 82, 93, 71
    ],
    'science_score': [
        95, 78, 100, 99, 83,
        88, 91, 74, 92, 96,
        69, 85, 90, 81, 79,
        94, 72, 87, 89, 97,
        77, 86, 93, 75, 80,
        70, 99, 84, 98, 73
    ]
}

df = pd.DataFrame(student_data)

df.head()
```

### ðŸ”¢ 2. Sorting Data

```python
# Sort by math_score
df.sort_values(by="math_score")
```

```python
# Sort by science_score descending
df.sort_values(by="science_score", ascending=False)
```

```python
# Sort by multiple columns
df.sort_values(by=["science_score", "math_score"], ascending=[False, True])
```

### ðŸ§ª 3. Filtering with Conditions

```python
# Who scored at least 90 in math?
df[df["math_score"] >= 90]
```

```python
# Who scored less than 70 in either subject?
df[(df["math_score"] < 70) | (df["science_score"] < 70)]
```

### ðŸ”— 4. Combining Filters

```python
# Students with B+ or higher in both
df[(df["math_score"] >= 87) & (df["science_score"] >= 87)]
```

### ðŸ§¾ 5. Using `.query()` for Readability

```python
# Same query using .query()
df.query("math_score >= 87 and science_score >= 87")
```

### ðŸ§ª Try It Yourself

- Sort the dataset by first name alphabetically
- Find students with the same math and science score
- Use `.query()` to find students who scored over 95 in *either* subject


```python
df_alphabetically = df.sort_values(by="first_name")
print(df_alphabetically)
```

```python
same_scores = df[df['math_score'] == df['science_score']]
print(same_scores)
```

```python
high_score = df.query("math_score > 95 or science_score > 95")
print(high_score)
```
