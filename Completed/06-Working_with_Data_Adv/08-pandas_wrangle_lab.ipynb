### ðŸ“¥ 1. Load the Dataset

Made my own data file since could not find students.csv file

```python
import pandas as pd
# df = pd.read_csv("../../data/students.csv")

student_data = {
    'first_name': [
        'Danielle', 'Angel', 'Joshua', 'Jeffrey', 'Jill',
        'Michael', 'Samantha', 'Brian', 'Olivia', 'Ethan',
        'Sophia', 'Ryan', 'Emma', 'David', 'Grace',
        'Lucas', 'Ava', 'Noah', 'Isabella', 'Mason',
        'Liam', 'Charlotte', 'Benjamin', 'Amelia', 'Elijah',
        'Harper', 'Alexander', 'Mia', 'James', 'Ella'
    ],
    'last_name': [
        'Wood', 'Clark', 'Adams', 'Zuniga', 'Wong',
        'Martinez', 'Taylor', 'Anderson', 'Nguyen', 'Hernandez',
        'Moore', 'King', 'Lopez', 'Scott', 'Green',
        'Turner', 'Hall', 'Allen', 'Baker', 'Nelson',
        'Carter', 'Mitchell', 'Perez', 'Roberts', 'Phillips',
        'Campbell', 'Evans', 'Parker', 'Murphy', 'Reed'
    ],
    'math_score': [
        100, 67, 61, 77, 75,
        89, 94, 72, 85, 91,
        68, 80, 97, 73, 84,
        90, 66, 79, 88, 92,
        76, 83, 95, 70, 87,
        64, 98, 82, 93, 71
    ],
    'science_score': [
        95, 78, 100, 99, 83,
        88, 91, 74, 92, 96,
        69, 85, 90, 81, 79,
        94, 72, 87, 89, 97,
        77, 86, 93, 75, 80,
        70, 99, 84, 98, 73
    ]
}

df = pd.DataFrame(student_data)
df.head()
```

### ðŸ”Ž 2. Inspect and Audit the Data

```python
# Basic overview
df.info()
```

```python
# Summary stats
df.describe()
```

### ðŸ§¼ 3. Clean Missing or Invalid Data

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop rows with missing names
df.dropna(subset=["first_name", "last_name"], inplace=True)

# Fill any missing scores with column average
df["math_score"].fillna(df["math_score"].mean(), inplace=True)
df["science_score"].fillna(df["science_score"].mean(), inplace=True)
```

### ðŸ§  4. Feature Engineering (New Columns)

```python
# Add average and grade
df["average_score"] = (df["math_score"] + df["science_score"]) / 2

def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "D"

df["grade"] = df["average_score"].apply(grade)
df.head()
```

### ðŸ”½ 5. Sorting and Filtering

```python
# Top performers
df[df["average_score"] > 90].sort_values(by="average_score", ascending=False).head()
```

### ðŸ“Š 6. Group and Describe by Grade

```python
# How many of each grade?
df["grade"].value_counts()
```

```python
# Average scores per grade group
df.groupby("grade")[["math_score", "science_score", "average_score"]].mean()
```

### ðŸ’¾ 7. Save the Cleaned Dataset

```python
df.to_csv("student_scores_cleaned.csv", index=False)
```

### ðŸ§  Challenge Task

> Your turn! Filter out students who got a D, sort by last name, and export to a new file:
- Only include columns: `first_name`, `last_name`, `grade`
- Save it as `"d_students.csv"`




```python
df["average_score"] = df[["math_score", "science_score"]].mean(axis=1)

def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'


df['grade'] = df['average_score'].apply(get_grade)
d_students = df.query('grade == "D"')
d_students = d_students.sort_values(by="last_name")

d_students = d_students[["first_name", "last_name", "grade"]]
print(d_students)
d_students.to_csv('d_students.csv', index=False)
```
