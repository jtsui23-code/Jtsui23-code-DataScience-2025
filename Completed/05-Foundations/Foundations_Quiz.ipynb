## Q4. File I/O
Write Python code that saves `Hello World` to a file called `hello.txt`, then reads it back and prints the contents.

```python
with open("hello.txt", "w") as f:
  f.write("Hello World")

with open("hello.txt", "r") as f:
  contents = f.read()
  print(contents)
```

## Q6. Plotting
Create a line plot of $y = x^2$ for $x$ values from 0 to 10.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11,1)
y = x**2

plt.plot(x,y, marker="o", color="blue", label="y = x^2")
plt.title("Line Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

## Q9. Timing
Compare the runtime of squaring numbers using a Python loop vs. NumPy. Use `%timeit` for each.

```python
import numpy as np

numbers = list(range(1000))
numbers_np = np.arange(1_000_000)


def square_loop(nums):
    result = []
    for x in nums:
        result.append(x**2)
    return result


def square_numpy(nums):
    return nums**2

%timeit square_loop(numbers)

%timeit square_numpy(numbers_np)


```
