## 1. Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

âœ… **Your Turn**: Plot `cos(x)` on the same graph with a different color and label.

```python
x = np.linspace(0,10, 100)
y = np.cos(x)

plt.plot(x, y, color="red", label="cos(x)")
plt.title("Cosine Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## 3. Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
```

âœ… **Your Turn**: Generate two sets of random points and plot them in different colors on the same scatter plot.

```python
x1 = np.random.rand(25)
y1 = np.random.rand(25)
size1 = 300 * np.random.rand(25)
color1 = "blue"

x2 = np.random.rand(25)
y2 = np.random.rand(25)
size2 = 300 * np.random.rand(25)
color2 = "yellow"

plt.scatter(x1, y1, s=size1,  c=color1, alpha=0.6)
plt.scatter(x2, y2, s=size2,  c=color2, alpha=0.6)

plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")

plt.colorbar(label="color scale")
plt.show()
```

## 4. Bar Plot

```python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

## 5. Customizing Plots

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

âœ… **Your Turn**: Customize your own plot with labels, legend, and grid lines.

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.tan(x)
y2 = np.sin(x)

plt.plot(x, y1, label="tan(x)", linestyle="-", color="purple")
plt.plot(x, y2, label="sin(x)", linestyle=":", color="green")

plt.title("Practice Customize Plot")
plt.xlabel("x (radian)")
plt.ylabel("value")
plt.ylim(-5, 5)

plt.legend()
plt.grid(True)
plt.show()
```
